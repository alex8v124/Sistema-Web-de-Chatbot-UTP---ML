```python
# sentiment_analysis.ipynb
# Asumimos que las variables analyzer de imports.ipynb y TextBlob están disponibles.

# Función para analizar sentimiento con VADER
def analyze_sentiment_vader(text):
    if analyzer is None or pd.isna(text): return 'Neutral' # Manejar valores NaN
    vs = analyzer.polarity_scores(text)
    if vs['compound'] >= 0.05:
        return 'Positivo'
    elif vs['compound'] <= -0.05:
        return 'Negativo'
    else:
        return 'Neutral'

# Función para analizar sentimiento con TextBlob (en español)
# Nota: TextBlob puede requerir traducción o un modelo específico para un mejor rendimiento en español
def analyze_sentiment_textblob(text):
    # TextBlob no tiene un estado que necesite ser inicializado como VADER's analyzer
    if pd.isna(text): return 'Neutral' # Manejar valores NaN
    blob = TextBlob(text)
    # Los umbrales pueden necesitar ajuste para español
    if blob.polarity > 0.1:
        return 'Positivo'
    elif blob.polarity < -0.1:
        return 'Negativo'
    else:
        return 'Neutral'

# Ejemplo de uso (para probar este notebook individualmente)
# test_query = "No me llegó el recibo de pago, estoy muy molesto."
# print(f"Test Query: '{test_query}' -> Sentimiento VADER: {analyze_sentiment_vader(test_query)}")
# print(f"Test Query: '{test_query}' -> Sentimiento TextBlob: {analyze_sentiment_textblob(test_query)}")
```
